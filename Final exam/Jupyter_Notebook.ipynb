<h3>Roll a couple of die:</h3>

```python
import random
def dice(n):
    x=random.randint(1,6)
    y=random.randint(1,6)
    for i in range(n):
        z=x+y
    return print('First dice:', x, 'Second dice:', y, 'Total points:', z)

dice(2)
```

<ul>

<li>The distance between Ad Confluentes and Rome, according to the Tabula Peutingeriana, is 201 Roman miles.</li>
<li>The distance from today's San Giovanni in Compito and the Fiumicino river is one Roman mile (1.48 km, 0.92 mi).</li>
